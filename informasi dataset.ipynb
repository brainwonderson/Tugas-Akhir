```python
# 13. Melihat distribusi nilai untuk kolom kategorikal
print("=== Distribusi Nilai Kategorikal ===")
for col in df.select_dtypes(include='object').columns:
    print(f"\nKolom: {col}")
    print(df[col].value_counts())
print("\n")
```
