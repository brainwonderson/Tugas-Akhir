```python
import pandas as pd

df1 = pd.read_csv("/content/drive/MyDrive/Dataset/train_transaction.csv")
df2 = pd.read_csv("/content/drive/MyDrive/Dataset/train_identity.csv")
df3 = pd.read_csv("/content/drive/MyDrive/Dataset/test_transaction.csv")
df4 = pd.read_csv("/content/drive/MyDrive/Dataset/test_identity.csv")
```

```python
# Merge train
train = df1.merge(df2, on="TransactionID", how="left")

# Merge test
test = df3.merge(df4, on="TransactionID", how="left")
```

```python
# Dalam format DataFrame agar lebih mudah dibaca
tipe_data_train_df = pd.DataFrame({
    "feature": train.columns,
    "dtype": train.dtypes.values
})

tipe_data_train_df
```

```python
# Tampilkan dalam DataFrame
total_non_null_df = pd.DataFrame({
    "feature": train.columns,
    "non_null_count": train.count().values,
    "missing_count": train.isnull().sum().values,
    "missing_percent": (train.isnull().mean().values * 100)
})

total_non_null_df
```

```python
# Dalam format tabel
tipe_data_test_df = pd.DataFrame({
    "feature": test.columns,
    "dtype": test.dtypes.values
})

tipe_data_test_df
```

```python
# Total data, missing count, dan missing percent pada test
total_non_null_test_df = pd.DataFrame({
    "feature": test.columns,
    "non_null_count": test.count().values,
    "missing_count": test.isnull().sum().values,
    "missing_percent": (test.isnull().mean().values * 100)
})

total_non_null_test_df
```

```python
# Mengecek jumlah dan persentase fraud pada dataset train

# Pastikan kolom isFraud ada
print("Apakah kolom 'isFraud' ada di train? ->", "isFraud" in train.columns)

# Jumlah tiap kelas (0 = non-fraud, 1 = fraud)
fraud_counts = train["isFraud"].value_counts()
print("\nJumlah transaksi per kelas:")
print(fraud_counts)

# Persentase tiap kelas
fraud_ratio = train["isFraud"].value_counts(normalize=True) * 100
print("\nPersentase transaksi per kelas (%):")
print(fraud_ratio)

# Optional: Imbalance Ratio (perbandingan non-fraud : fraud)
IR = fraud_counts[0] / fraud_counts[1]
print(f"\nImbalance Ratio (IR) = {IR:.2f} (non-fraud : fraud)")
```
